```python
import tensorflow as tf
```

```python

from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size=10000
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python

import numpy as np
```

```python

sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
# To use pad sequences make sure size of the sentences is same
# S for this we are using padding 'pre' so that it will add 0s where words are not present in the sentences
#this will make the sent_length or sentence length to be 8 a mentioned
```

```python
dim=10
```

```python

model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.summary()
```

```python

print(model.predict(embedded_docs))
```
